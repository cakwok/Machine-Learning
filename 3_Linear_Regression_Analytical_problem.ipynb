```python
import pandas as pd
import numpy as np
```

```python
d={"X1": [0,0,1,1], "X2":[0,1,0,1], "Y": [0, 1.5, 2, 2.5]}    #Setup dataset
dataset = pd.DataFrame(data = d)
dataset.head()
```

```python
feature_cols = ['X1', 'X2']
x = dataset[feature_cols]                                     #assign columns as x features
y = dataset['Y']                                              #assign column as y
```

<p>As discussed with professor, use normal equation is accepted as it is the same as least square linear regression.  

Therefore, thetas are come up by (x<sup>T</sup>x)<sup>-1</sup>x<sup>T</sup>y</p>

```python
x_with_bias = np.c_[ np.ones(x.shape[0]), x]                       #prepend a column of all 1 as bias
print("feature columns with bias", "\n", x_with_bias, "\n")

theta = np.linalg.inv(x_with_bias.T.dot(x_with_bias)).dot(x_with_bias.T).dot(y) #compute theta by normal equation
print("thetas are", theta)
```
