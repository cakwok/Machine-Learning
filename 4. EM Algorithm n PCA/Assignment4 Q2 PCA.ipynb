```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from numpy import linalg as l_algebra                   #Q2.3 linalg is linear algebra, to compute eigen pairs
from sklearn.linear_model import LogisticRegression     #Q2.4 import logistic regressor
from sklearn.metrics import confusion_matrix            #Q2.4 to compare PCA performance
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from sklearn import metrics                             #Q2.5 compute PCA performance
```

```python
#EDA
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/CS6140 Assignment4/dataset.csv")
df.head(10)
```

```python
#Q2.1 Standardize dataset
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

X_standardized = StandardScaler().fit_transform(X)
print("Standardized dataset:")
print(X_standardized)
```

```python
#Q2.  Split dataset and apply eigen decomposition
X_train, X_test, y_train, y_test = train_test_split(X_standardized , y, test_size=0.2, random_state=42)   # 80% training and 20% testing

X_train_mean = np.mean(X_train, axis=0)
covariance__matrix = (X_train - X_train_mean).T.dot((X_train - X_train_mean)) / (X_train.shape[0] - 1)

eigenvalues, eigenvectors = l_algebra.eig(covariance__matrix)

print("Eigenvalues: \n", eigenvalues, "\n")
print("Eigvenvectors: \n", eigenvectors, "\n")
print("Number of Eigenvalues: ", len(eigenvalues))
```

```python
#Q2.3 Select first 6 principal components with largest eigenvalues
explained_variance = [eigenvalue / np.sum(eigenvalues) for eigenvalue in sorted(eigenvalues, reverse = True)]

print("First 6 principal components in explained_variance: \n", explained_variance[:6], "\n")

principal_components = [(np.abs(eigenvalues[i].real), eigenvectors[:, i].real) for i in range(len(eigenvalues))]
principal_components.sort(key=lambda x: x[0], reverse=True)

print("First 6 principal components: (Eigenvalues, Eigenvectors)")

for i in range (6):
  print(principal_components[i])
```

```python
#Q2.4. Project training set into the new 6 dimensional feature space

projection_matrix = np.hstack((principal_components[0][1].reshape(754,1), 
                               principal_components[1][1].reshape(754,1),
                               principal_components[2][1].reshape(754,1),
                               principal_components[3][1].reshape(754,1),
                               principal_components[4][1].reshape(754,1),
                               principal_components[5][1].reshape(754,1),))

print("projection_matrix.shape", projection_matrix.shape, "\n")
print("projection_matrix \n", projection_matrix, "\n")

projected_X_train = X_train.dot(projection_matrix)      #project all data points by projection matrix

print("projected_X_train.shape", projected_X_train.shape)
```

```python
#Q2.5 Train two logistic regression models and report ROC curves

classifier_logistic = LogisticRegression(random_state = 42, max_iter=200)
classifier_logistic.fit(X_train, y_train)

classifier_my_pca = LogisticRegression(random_state=42, max_iter=200)
classifier_my_pca.fit(projected_X_train, y_train)

y_pred = classifier_logistic.predict(X_train) 
y_pred2 = classifier_my_pca.predict(projected_X_train) 

metrics.plot_roc_curve(classifier_logistic, X_train, y_train)
plt.title("ROC Curve for standardized training dataset") 
plt.show()

metrics.plot_roc_curve(classifier_my_pca, projected_X_train, y_train) 
plt.title("ROC Curve for projected training dataset") 
plt.show()
```

```python
#Q2.6 Validate the model using testing dataset

projected_X_test = X_test.dot(projection_matrix)    # Project testing dataset into feature space

y_pred = classifier_logistic.predict(X_test) 
y_pred2 = classifier_my_pca.predict(projected_X_test) 

print("Predicted target variable by 6 dimensional feature space: \n", y_pred2)
```

```python
#Q2.7 Compare the logistic regression models before and after PCA 

print("True label distribution: y = 0 counts:", df['class'].value_counts()[0])
print("True label distribution: y = 1 counts:", df['class'].value_counts()[1], "\n")

print("Logistic regressions before PCA:")
print("-------------------------------- \n")
print('Accuracy:', accuracy_score(y_test, y_pred), "\n")
print('Confusion Matrix: \n',  confusion_matrix(y_test, y_pred), "\n")
print('Classification Report: \n', classification_report(y_test, y_pred), "\n")

print("Logistic regressions after PCA:")
print("------------------------------- \n")
print('Accuracy:', accuracy_score(y_test, y_pred2), "\n")
print('Confusion Matrix: \n',  confusion_matrix(y_test, y_pred2), "\n")
print('Classification Report: \n', classification_report(y_test, y_pred2), "\n")

metrics.plot_roc_curve(classifier_logistic, X_test, y_test) 
plt.title("ROC Curve for Logistic regressions before PCA")
plt.show()

metrics.plot_roc_curve(classifier_my_pca, projected_X_test, y_test) 
plt.title("ROC Curve for Logistic regressions after PCA")
plt.show()
```

The results shows similar ROC curves, but AUC for label value 1 drops after applying PCA .

As shown with the analysis of dataset imbalance above, true labels in the dataset are prone to value 1.  When comparing the ROC curves and accuracy table, it looks better result with precision and recall for value 0 after PCA.

However, PCA and logistic regressions are 2 different techniques and it may not carry meanings when compared.  PCA aims at decorrelate dependent variables, so when prediction are modelled based on the assumption of independent variables, PCA could help to decorrelate dependence, resulting the prediction models more reliable, and also less datapoints to train.  

However, there is no direct correlation with better prediction. Ultimate improvement depends on how existing variables are already independent.  If they are already independent, after PCA it might not shows significance;  if they are not, PCA would show improvement.
